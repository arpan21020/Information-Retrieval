```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('cleaned_tweets.csv')

```

```python
# Check for null values
null_values = df.isnull().sum()
print(null_values)
```

```python

plt.figure(figsize=(10, 6))

df=df.sample(100)
plt.scatter(df['Id'], df['Views Count'])
plt.xticks(rotation=90)
plt.xlabel('Tweets Id')
plt.ylabel('Views Count')
plt.show()
```

```python
# # Generate a histogram for a column
df['Policy Name'].hist()
plt.xticks(rotation='vertical')
plt.xlabel('Policies')
plt.ylabel('Tweets Count')
plt.show()
```

```python
# Generate a bar graph for a column
df['Language'].value_counts().plot.bar()
plt.xticks(rotation=90)
plt.xlabel('Languages')
plt.ylabel('Number of tweets')
plt.show()
```

```python
df=df.head(1000)
df.plot.scatter(x='Id', y='Retweet Count')
plt.xticks(rotation=90)
plt.xlabel('Tweets Id')
plt.ylabel('Retweet Count')
plt.show()
```

```python
# Generate a scatter plot for two columns
df.plot.scatter(x='Id', y='User Followers Count')
plt.xticks(rotation=90)
plt.xlabel('Tweets Id')
plt.ylabel('Followers Count on that Tweet')
plt.show()
```
